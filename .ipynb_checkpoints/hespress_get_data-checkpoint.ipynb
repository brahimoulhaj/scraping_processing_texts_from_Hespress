```python
import requests
import re
from scrapy.selector import Selector
```

```python
archives = [
    'https://www.hespress.com/archive/2019/9',
    'https://www.hespress.com/archive/2019/10'
]

def get_pages_links_from_archive(archive_link):
    body = requests.get(archive_link)
    total_posts = Selector(text = body.text).css('div#box_pagination span.pagination_total::text').get()
    total_posts = 1+(int(total_posts)//10)
    pages_links = []
    for i in range(1,total_posts+1):
        pages_links.append('{}/index.{}.html'.format(archive_link, i))
    return pages_links

def get_posts_links_from_page(page_link):
    body = requests.get(page_link)
    posts_links = Selector(text=body.text).css('div.short div.short_holder h2.section_title a::attr(href)').getall()
    posts_links = ['https://www.hespress.com'+l for l in posts_links]
    return posts_links

def get_post_comments(post_link):
    body = requests.get(post_link)
    comments = Selector(text=body.text).css('div.comment_body_in div.comment_text::text').getall()
    return comments
```

```python
%%time
pages_links = []
for a in archives: pages_links.extend(get_pages_links_from_archive(a))
print(len(pages_links))
```

```python
%%time
posts_links = []
for link in pages_links: posts_links.extend(get_posts_links_from_page(link))
```

```python
%%time
import threading
posts_comments = []

def thread_func(posts):
    for p in posts: posts_comments.extend(get_post_comments(p))
        
thread_1 = threading.Thread(target=thread_func, args=(posts_1,))
thread_1.start()

thread_2 = threading.Thread(target=thread_func, args=(posts_2,))
thread_2.start()

thread_3 = threading.Thread(target=thread_func, args=(posts_3,))
thread_3.start()

thread_4 = threading.Thread(target=thread_func, args=(posts_4,))
thread_4.start()

thread_1.join()
thread_2.join()
thread_3.join()
thread_4.join()

print(len(posts_comments))
```

```python
comments = str(posts_comments).strip('[]')
comments = re.sub('[^ء-يﭐ-﷿ﹰ-ﻼ]',' ',comments)
comments = re.sub(' +',' ',comments)
print(len(comments))
```

```python
f = open("hespress.txt", "a", encoding="utf-8")
f.write(comments)
f.close()
```
